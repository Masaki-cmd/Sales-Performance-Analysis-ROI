## データの読み込みと初期クレンジング

### Subtask:
解凍されたCSVファイル 'Online-eCommerce.csv' をpandas DataFrameとして読み込み、データ型、欠損値、重複を確認し、必要に応じて基本的なデータクレンジングを実施します。


**Reasoning**:
The first step for 'データの読み込みと初期クレンジング' is to import pandas, load the CSV into a DataFrame, and then display its head, info, missing values, and duplicates to get an initial understanding of the data.



````python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = '/content/extracted_data/Online-eCommerce.csv'
df = pd.read_csv(file_path)

# Display the first 5 rows of the DataFrame
print("\n--- First 5 rows of the DataFrame ---")
print(df.head())

# Display basic information about the DataFrame
print("\n--- DataFrame Info ---")
df.info()

# Display the total number of missing values per column
print("\n--- Missing values per column ---")
print(df.isnull().sum())

# Display the number of duplicate rows
print("\n--- Number of duplicate rows ---")
print(df.duplicated().sum())
````

**Reasoning**:
The previous step identified 14 duplicate rows and 15 rows with missing values across all columns. The next action is to remove the duplicate rows first, as this is a straightforward cleaning step, and then drop the rows with missing values, which appears to be a systemic issue (all columns have the same number of nulls).



````python
print("\n--- Cleaning Data ---")

# Remove duplicate rows
df.drop_duplicates(inplace=True)
print(f"Removed duplicate rows. New DataFrame shape: {df.shape}")

# Drop rows with any missing values
df.dropna(inplace=True)
print(f"Removed rows with missing values. New DataFrame shape: {df.shape}")

# Verify no missing values
print("\n--- Missing values after cleaning ---")
print(df.isnull().sum().sum())

# Verify no duplicate rows
print("\n--- Duplicate rows after cleaning ---")
print(df.duplicated().sum())
````

**Reasoning**:
Convert the 'Order_Date' column to datetime objects and 'Order_Number' to integer type for proper temporal analysis and identification. Then, display the DataFrame's info to verify the data type changes.



````python
print("\n--- Converting data types ---")

df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Order_Number'] = df['Order_Number'].astype(int)

# Display basic information about the DataFrame again to verify data types
print("\n--- DataFrame Info after type conversion ---")
df.info()
````

**Reasoning**:
The previous code failed because the 'Order_Date' column has a date format of 'DD/MM/YYYY', which is not the default 'MM/DD/YYYY' that `pd.to_datetime` expects. I will explicitly specify the correct format `"%d/%m/%Y"` to correctly convert the 'Order_Date' column to datetime objects.



````python
print("\n--- Converting data types ---")

df['Order_Date'] = pd.to_datetime(df['Order_Date'], format="%d/%m/%Y")
df['Order_Number'] = df['Order_Number'].astype(int)

# Display basic information about the DataFrame again to verify data types
print("\n--- DataFrame Info after type conversion ---")
df.info()
````

## カテゴリ名の統一と利益関連指標の計算

### Subtask:
「Motherboard」と「MotherBoard」の表記ゆれを修正し、「Profit」と「Profit_Margin」カラムを計算して追加します。これにより、製品カテゴリごとの正確な収益分析の基盤を確立します。


**Reasoning**:
The first step is to correct the typo in the 'Category' column by replacing 'MotherBoard' with 'Motherboard'. This will ensure data consistency for subsequent analysis.



````python
df['Category'] = df['Category'].replace('MotherBoard', 'Motherboard')

print("Category column after typo correction:")
print(df['Category'].value_counts())
````

**Reasoning**:
The previous step corrected the 'Category' typo. Now, as per the subtask, calculate the 'Profit' and 'Profit_Margin' columns. Profit is `Total_Sales - Total_Cost`, and Profit Margin is `(Profit / Total_Sales) * 100`.



````python
df['Profit'] = df['Total_Sales'] - df['Total_Cost']
df['Profit_Margin'] = (df['Profit'] / df['Total_Sales']) * 100

print("DataFrame with new 'Profit' and 'Profit_Margin' columns:")
print(df.head())
````

## 商品・カテゴリ・ブランド別利益分析

### Subtask:
`Product`、`Category`、`Brand`ごとに総売上、総コスト、総利益、平均利益率を計算し、利益率が低い商品やカテゴリを特定します。この分析結果を可視化し、収益性の課題を明確にします。


**Reasoning**:
As per the instructions, the first step is to aggregate the data by 'Category' and calculate the total sales, total cost, total profit, and average profit margin for each category. This will give an initial overview of the profitability by category.



````python
category_profit_analysis = df.groupby('Category').agg(
    total_sales=('Total_Sales', 'sum'),
    total_cost=('Total_Cost', 'sum'),
    total_profit=('Profit', 'sum'),
    average_profit_margin=('Profit_Margin', 'mean')
).sort_values(by='average_profit_margin', ascending=True)

print("\n--- Category-wise Profit Analysis (Sorted by Average Profit Margin) ---")
print(category_profit_analysis)
````

**Reasoning**:
The previous step analyzed profit by 'Category'. Now, as per the subtask instructions, I will aggregate the data by 'Brand' to calculate total sales, total cost, total profit, and average profit margin for each brand. This will help identify brands with potentially lower profitability.



````python
brand_profit_analysis = df.groupby('Brand').agg(
    total_sales=('Total_Sales', 'sum'),
    total_cost=('Total_Cost', 'sum'),
    total_profit=('Profit', 'sum'),
    average_profit_margin=('Profit_Margin', 'mean')
).sort_values(by='average_profit_margin', ascending=True)

print("\n--- Brand-wise Profit Analysis (Sorted by Average Profit Margin) ---")
print(brand_profit_analysis)
````

**Reasoning**:
The previous steps calculated profit analysis for 'Category' and 'Brand'. Now, to fulfill the subtask, I need to aggregate the data by 'Product' to calculate total sales, total cost, total profit, and average profit margin for each product.



````python
product_profit_analysis = df.groupby('Product').agg(
    total_sales=('Total_Sales', 'sum'),
    total_cost=('Total_Cost', 'sum'),
    total_profit=('Profit', 'sum'),
    average_profit_margin=('Profit_Margin', 'mean')
).sort_values(by='average_profit_margin', ascending=True)

print("\n--- Product-wise Profit Analysis (Sorted by Average Profit Margin) ---")
print(product_profit_analysis)
````

**Reasoning**:
I have already calculated the profit analysis for categories, brands, and products. The next step is to visualize these results to clearly identify profitability issues as stated in the subtask instructions. I will start by visualizing the category-wise profit analysis using a bar chart for total profit.



````python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize Category-wise Profit Analysis
plt.figure(figsize=(12, 6))
sns.barplot(x=category_profit_analysis.index, y='total_profit', data=category_profit_analysis)
plt.title('Total Profit by Category')
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=category_profit_analysis.index, y='average_profit_margin', data=category_profit_analysis)
plt.title('Average Profit Margin by Category')
plt.xlabel('Category')
plt.ylabel('Average Profit Margin (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

**Reasoning**:
The previous step visualized the category-wise profit analysis. Now, to continue visualizing the profit analysis by different dimensions as required by the subtask, I will visualize the brand-wise total profit and average profit margin using bar charts.



````python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize Brand-wise Profit Analysis - Total Profit
plt.figure(figsize=(12, 6))
sns.barplot(x=brand_profit_analysis.index, y='total_profit', data=brand_profit_analysis)
plt.title('Total Profit by Brand')
plt.xlabel('Brand')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualize Brand-wise Profit Analysis - Average Profit Margin
plt.figure(figsize=(12, 6))
sns.barplot(x=brand_profit_analysis.index, y='average_profit_margin', data=brand_profit_analysis)
plt.title('Average Profit Margin by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Profit Margin (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

**Reasoning**:
The next step is to visualize the product-wise profit analysis, specifically focusing on the top and bottom products by total profit to identify areas for revenue enhancement or cost optimization. This directly addresses the visualization requirement of the subtask.



````python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort products by total profit to identify top and bottom performers
product_profit_sorted_by_profit = product_profit_analysis.sort_values(by='total_profit', ascending=False)

# Visualize Top 10 Products by Total Profit
plt.figure(figsize=(14, 7))
sns.barplot(x=product_profit_sorted_by_profit['total_profit'].head(10).index, y=product_profit_sorted_by_profit['total_profit'].head(10).values)
plt.title('Top 10 Products by Total Profit')
plt.xlabel('Product')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualize Bottom 10 Products by Total Profit
plt.figure(figsize=(14, 7))
sns.barplot(x=product_profit_sorted_by_profit['total_profit'].tail(10).index, y=product_profit_sorted_by_profit['total_profit'].tail(10).values)
plt.title('Bottom 10 Products by Total Profit')
plt.xlabel('Product')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

## 顧客セグメンテーション分析

### Subtask:
顧客名ごとに総購入額、購入回数を計算し、優良顧客を特定します。これにより、パーソナライズされたマーケティング戦略の立案に役立つ洞察を得ます。「なぜこのグラフを作ったのか」「この結果からなぜその提案が導かれるのか」という論理の飛躍をなくすプロセスをGoogle Colabのテキストセルに詳しく書いてください。


**Reasoning**:
I need to group the DataFrame by `Customer_Name` and calculate the sum of `Total_Sales` and the count of `Order_Number` to create the `customer_segmentation` DataFrame, as per the first instruction.



````python
customer_segmentation = df.groupby('Customer_Name').agg(
    total_purchase_amount=('Total_Sales', 'sum'),
    purchase_count=('Order_Number', 'count')
).sort_values(by='total_purchase_amount', ascending=False)

print("--- Customer Segmentation DataFrame (Sorted by Total Purchase Amount) ---")
print(customer_segmentation.head())
````

**Reasoning**:
The previous step calculated the customer segmentation data. To identify valuable customers and gain insights for personalized marketing, I need to visualize the distribution of `total_purchase_amount` and `purchase_count`. I will start by visualizing the top 10 customers based on `total_purchase_amount` and `purchase_count` using bar plots.



````python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort by total purchase amount to get top customers
top_customers_by_amount = customer_segmentation.sort_values(by='total_purchase_amount', ascending=False).head(10)

# Plot Top 10 Customers by Total Purchase Amount
plt.figure(figsize=(12, 6))
sns.barplot(x=top_customers_by_amount.index, y='total_purchase_amount', data=top_customers_by_amount)
plt.title('Top 10 Customers by Total Purchase Amount')
plt.xlabel('Customer Name')
plt.ylabel('Total Purchase Amount')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Sort by purchase count to get top frequent customers
top_customers_by_count = customer_segmentation.sort_values(by='purchase_count', ascending=False).head(10)

# Plot Top 10 Customers by Purchase Count
plt.figure(figsize=(12, 6))
sns.barplot(x=top_customers_by_count.index, y='purchase_count', data=top_customers_by_count)
plt.title('Top 10 Customers by Purchase Count')
plt.xlabel('Customer Name')
plt.ylabel('Purchase Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

`State_Code`ごとに総売上と総利益を計算し、地域間のパフォーマンスのばらつきを分析します。パフォーマンスの低い地域を特定し、ターゲットを絞った戦略の可能性を探ります。

Then, I will sort the resulting DataFrame by `total_sales` in descending order and print its head.

````python
supervisor_performance = df.groupby('Assigned Supervisor').agg(
    order_count=('Order_Number', 'count'),
    total_sales=('Total_Sales', 'sum'),
    total_profit=('Profit', 'sum')
).sort_values(by='total_sales', ascending=False)

print("--- Supervisor Performance Analysis (Sorted by Total Sales) ---")
print(supervisor_performance)
````

To visually compare the performance of each supervisor, I will create three bar plots: one for total sales, one for total profit, and one for order count. This will allow for easy identification of high-performing and low-performing supervisors across these key metrics.



````python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize Supervisor Performance - Total Sales
plt.figure(figsize=(10, 6))
sns.barplot(x=supervisor_performance.index, y='total_sales', data=supervisor_performance, palette='viridis')
plt.title('Total Sales by Assigned Supervisor')
plt.xlabel('Assigned Supervisor')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualize Supervisor Performance - Total Profit
plt.figure(figsize=(10, 6))
sns.barplot(x=supervisor_performance.index, y='total_profit', data=supervisor_performance, palette='viridis')
plt.title('Total Profit by Assigned Supervisor')
plt.xlabel('Assigned Supervisor')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualize Supervisor Performance - Order Count
plt.figure(figsize=(10, 6))
sns.barplot(x=supervisor_performance.index, y='order_count', data=supervisor_performance, palette='viridis')
plt.title('Order Count by Assigned Supervisor')
plt.xlabel('Assigned Supervisor')
plt.ylabel('Order Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

Then, I will group the DataFrame by these new columns to calculate the total sales and total profit for each month and year, storing the result in a new DataFrame called `monthly_trends`.



````python
df['Order_Year'] = df['Order_Date'].dt.year
df['Order_Month'] = df['Order_Date'].dt.month

monthly_trends = df.groupby(['Order_Year', 'Order_Month']).agg(
    total_sales=('Total_Sales', 'sum'),
    total_profit=('Profit', 'sum')
).reset_index()

# Sort by year and month to ensure correct chronological order
monthly_trends = monthly_trends.sort_values(by=['Order_Year', 'Order_Month'])

print("--- Monthly Sales and Profit Trends ---")
print(monthly_trends.head())
print(monthly_trends.tail())
````

同じ注文番号（Order_Number）内で一緒に購入される頻度の高い商品を特定します。この分析を通じて、製品のバンドル販売やクロスセル戦略の機会を特定し、関連商品のプロモーションを強化するための洞察を得ます。


**Reasoning**:
To perform product co-occurrence analysis, I first need to group the products by their 'Order_Number' to create 'baskets' of items for each transaction. This is the initial step for association rule mining.



````python
import pandas as pd

# Group products by Order_Number to create transactional data
transactions = df.groupby('Order_Number')['Product'].apply(list).reset_index()
transactions.columns = ['Order_Number', 'Products']

print("--- Sample of transactional data (products per order) ---")
print(transactions.head())
````

**Reasoning**:
The previous step prepared the transactional data. Now, to perform association rule mining, I need to convert the list of products in each transaction into a one-hot encoded format. This is a necessary preprocessing step before applying algorithms like Apriori.



````python
from mlxtend.preprocessing import TransactionEncoder

# Convert the list of products into a one-hot encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit(transactions['Products']).transform(transactions['Products'])
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

print("--- Sample of one-hot encoded transactional data ---")
print(df_encoded.head())
````

**Reasoning**:
The previous steps prepared the one-hot encoded transactional data. Now, to identify frequently purchased items together, I need to apply the Apriori algorithm on the `df_encoded` DataFrame to find frequent itemsets. This is a core step in association rule mining for product co-occurrence analysis.



````python
from mlxtend.frequent_patterns import apriori

# Find frequent itemsets using the Apriori algorithm
# Using a min_support of 0.01 (1%) to find items that appear in at least 1% of transactions
frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)

# Sort frequent itemsets by support in descending order
frequent_itemsets = frequent_itemsets.sort_values(by='support', ascending=False)

print("--- Top 10 Frequent Itemsets ---")
print(frequent_itemsets.head(10))
````

**Reasoning**:
The previous step identified frequent itemsets. To understand product co-occurrence for bundle sales and cross-selling, I need to generate association rules from these frequent itemsets. I will use the `association_rules` function, setting `metric="lift"` and `min_threshold=1` to find rules with a positive correlation between items.



````python
from mlxtend.frequent_patterns import association_rules

# Generate association rules from the frequent itemsets
# Using 'lift' as the metric to find rules that occur more often than expected by chance
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Sort the rules by lift in descending order to see the strongest associations
rules = rules.sort_values(by='lift', ascending=False)

print("--- Top 10 Association Rules (Sorted by Lift) ---")
print(rules.head(10))
````

**Reasoning**:
To find more frequent itemsets and enable the generation of association rules, I need to lower the `min_support` value in the `apriori` algorithm. I will set it to 0.001 and then generate the association rules again with the `lift` metric and a `min_threshold` of 1.



````python
from mlxtend.frequent_patterns import apriori, association_rules

# Find frequent itemsets using the Apriori algorithm with a lower min_support
frequent_itemsets = apriori(df_encoded, min_support=0.001, use_colnames=True)

# Sort frequent itemsets by support in descending order
frequent_itemsets = frequent_itemsets.sort_values(by='support', ascending=False)

print("--- Top 10 Frequent Itemsets (with min_support=0.001) ---")
print(frequent_itemsets.head(10))

# Generate association rules from the frequent itemsets
# Using 'lift' as the metric to find rules that occur more often than expected by chance
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Sort the rules by lift in descending order to see the strongest associations
rules = rules.sort_values(by='lift', ascending=False)

print("\n--- Top 10 Association Rules (Sorted by Lift) ---")
print(rules.head(10))
````

````python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Convert the list of categories into a one-hot encoded DataFrame
te_cat = TransactionEncoder()
te_ary_cat = te_cat.fit(category_transactions['Categories']).transform(category_transactions['Categories'])
df_encoded_cat = pd.DataFrame(te_ary_cat, columns=te_cat.columns_)

print("--- Sample of one-hot encoded category data ---")
print(df_encoded_cat.head())

# Find frequent itemsets using the Apriori algorithm for categories
# Starting with a min_support of 0.01
frequent_itemsets_cat = apriori(df_encoded_cat, min_support=0.01, use_colnames=True)

# Sort frequent itemsets by support in descending order
frequent_itemsets_cat = frequent_itemsets_cat.sort_values(by='support', ascending=False)

print("\n--- Top 10 Frequent Category Itemsets ---")
print(frequent_itemsets_cat.head(10))

# Generate association rules from the frequent category itemsets
# Using 'lift' as the metric to find rules that occur more often than expected by chance
rules_cat = association_rules(frequent_itemsets_cat, metric="lift", min_threshold=1)

# Sort the rules by lift in descending order to see the strongest associations
rules_cat = rules_cat.sort_values(by='lift', ascending=False)

print("\n--- Top 10 Association Rules for Categories (Sorted by Lift) ---")
print(rules_cat.head(10))
````
